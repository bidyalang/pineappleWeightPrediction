```python
import os
import cv2
import albumentations as A
from tqdm import tqdm

# Define augmentation
transform = A.Compose([
    A.RandomBrightnessContrast(p=0.5),
    A.GaussianNoise(p=0.5),
    A.Rotate(limit=20, p=0.5)
], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))

# Base paths
base_path = '/content/drive/MyDrive/detection'  # change this if your path is different
image_base = os.path.join(base_path, 'images')
label_base = os.path.join(base_path, 'labels')
output_image_base = os.path.join(base_path, 'augmented/images')
output_label_base = os.path.join(base_path, 'augmented/labels')

# Split folders
splits = ['train', 'val', 'test']

# Loop through each split
for split in splits:
    image_dir = os.path.join(image_base, split)
    label_dir = os.path.join(label_base, split)
    out_img_dir = os.path.join(output_image_base, split)
    out_lbl_dir = os.path.join(output_label_base, split)

    os.makedirs(out_img_dir, exist_ok=True)
    os.makedirs(out_lbl_dir, exist_ok=True)

    for filename in tqdm(os.listdir(image_dir), desc=f"Augmenting {split} set"):
        if not filename.endswith(('.jpg', '.png')):
            continue

        name, ext = os.path.splitext(filename)
        img_path = os.path.join(image_dir, filename)
        label_path = os.path.join(label_dir, name + ".txt")

        # Load image
        image = cv2.imread(img_path)
        if image is None:
            continue
        h, w = image.shape[:2]

        # Load labels
        bboxes, class_labels = [], []
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                for line in f.readlines():
                    parts = line.strip().split()
                    if len(parts) == 5:
                        cls, x, y, bw, bh = parts
                        bboxes.append([float(x), float(y), float(bw), float(bh)])
                        class_labels.append(cls)

        # Apply augmentation
        try:
            transformed = transform(image=image, bboxes=bboxes, class_labels=class_labels)
        except Exception as e:
            print(f"Skipped {filename} due to error: {e}")
            continue

        # Save new image
        new_img = transformed['image']
        new_img_name = f"aug_{filename}"
        cv2.imwrite(os.path.join(out_img_dir, new_img_name), new_img)

        # Save new labels
        new_label_path = os.path.join(out_lbl_dir, f"aug_{name}.txt")
        with open(new_label_path, 'w') as f:
            for bbox, cls in zip(transformed['bboxes'], class_labels):
                f.write(f"{cls} {' '.join(f'{x:.6f}' for x in bbox)}\n")
```

```python
import torch
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import os

# Load model
model = YOLO('/content/drive/MyDrive/train2/weights/best.pt')

# Load test data
test_dir = '/content/drive/MyDrive/Classification/molvumClassification/images/test'
class_names = sorted(os.listdir(test_dir))

# Collect true labels and predictions
true_labels = []
pred_labels = []

for class_index, class_name in enumerate(class_names):
    class_path = os.path.join(test_dir, class_name)
    for img_file in os.listdir(class_path):
        img_path = os.path.join(class_path, img_file)
        results = model(img_path, verbose=False)
        pred = results[0].probs.top1
        true_labels.append(class_index)
        pred_labels.append(pred)

# Print per-class report
print(classification_report(true_labels, pred_labels, target_names=class_names))
```
